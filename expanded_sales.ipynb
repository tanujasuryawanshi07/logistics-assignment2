```python
import pandas as pd
```

```python
df = pd.read_csv('expanded_sales.csv')
print(df.head())
```

```python
# Selecting a single cell by row index and column name
print("Row 3, Region:", df.loc[3, "Region"])  # row number , column label / column label list
```

```python
# Selecting multiple specific columns for a specific row
print("\nRow 3 Info:\n", df.loc[3, ["Region", "Revenue"]])
```

```python
print("\nRows 1 to 3, columns Product through Orders:\n", df.loc[1:3, "Product":"Orders"])   # start : end
```

```python
print("\nCancelled Products:\n", df.loc[df["Delivery_Status"] == "Cancelled", "Product"])
```

```python
 
# Row position 0, Column position 3 (Product)
print("Row 0, Column 3 Value:", df.iloc[0, 3])
```

```python
print("\nSlice rows 1-3 and columns 1-3:\n", df.iloc[1:4, 1:4])
```

```python
# Selecting specific isolated positions using lists of numbers
# Grab row positions 0 and 5, and column positions 3 and 4 (Product and Revenue)
print("\nSpecific row/column pairs:\n", df.iloc[[0, 5], [3, 4]])
```

```python
# Grabbing everything up to the end using shortcuts
# Leaving the row side blank [:] means ALL rows. [:-1] means all columns except the last one.
print("\nAll rows, all columns except the last one:\n", df.iloc[2:-1, 3:-1])
```

```python
# Creating a column using cross-column math
# Calculate the Average Value per Order
df["Value_Per_Order"] = df["Revenue"] / df["Orders"]
print(df[["Product", "Revenue", "Orders", "Value_Per_Order"]])
```

```python
# Creating a column via flat broadcasting  
# Let's say there is a flat delivery surcharge of $15 added to every order
df["Surcharge"] = 7
print(df.head())
```

```python
# Overwriting a column conditionally using .loc
# If a row has Delivery_Status as "Cancelled", let's force its Revenue to 0!
# Syntax: df.loc[condition, column_to_update] = new_value
df.loc[df["Delivery_Status"] == "Cancelled", "Revenue"] = 0

df["Value_Per_Order"] = df["Revenue"] / df["Orders"]

print(df[["Product", "Revenue", "Delivery_Status", "Value_Per_Order", "Surcharge"]].head())
```

```python
cat_revenue = df.groupby("Category")["Revenue"].sum()
print("Total Revenue by Category:\n", cat_revenue)
```

```python
regional_cat_sales = df.groupby(["Region", "Category"])["Revenue"].sum()
print("\nRevenue by Region & Category:\n", regional_cat_sales)
```

```python
custom_pipeline = df.groupby("Region").agg({
    "Revenue": ["sum", "max"],  # Finds both the total sum and the highest transaction value
    "Orders": ["mean", "sum"]   # Finds the average number of orders
})
print("\nCustom Aggregation Pipeline:\n", custom_pipeline)
```
