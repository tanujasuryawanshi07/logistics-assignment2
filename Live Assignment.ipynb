```python
import numpy as np

print("--- 1. Simple Slicing ---")
scores = np.array([10, 25, 70,65, 55,17])

print("5 scores:", scores[1:4])   # start , end-1

print("\n--- 2. Filtering with Conditions ---")
mask = scores <= 55     
print("True/False Mask:", mask)


passed_scores = scores[mask]
print("Actual Passing Scores:", passed_scores)
```
