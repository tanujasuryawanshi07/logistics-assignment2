```python
salaries = []  #Empty list create

with open('salaries.csv', 'r') as file:   #salaries.csv --> file varible
    lines = file.readlines()
    for line in lines[1:]:     #[1:]  
        parts = line.strip().split(',')   #Eva,Data Science,120000   ->   parts = ['Eva', 'Data Science', '120000']
        salary = int(parts[2])
        salaries.append(salary)  #salaries []   ->  salaries = [120000,90000,75000,110000,95000]

total_salary = sum(salaries) # sum => total_salary(300000)
num_employees = len(salaries)   
average_salary = total_salary / num_employees


print(f"--- Salary Statistics ---")
print(f"Total Employees: {num_employees}")
print(f"Total Payroll:  ${total_salary:,}")
print(f"Average Salary: ${average_salary:,.2f}")
```
